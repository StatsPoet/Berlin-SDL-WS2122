## create a df for brightness where every missing value is filled in with the mean. Should prevent a bias due to missing pictures

```python
full_brightness_df_mean = full_brightness_df.copy()
```

```python
full_brightness_df_mean= full_brightness_df_mean.replace(np.nan, 0)
```

```python
for j in range(full_brightness_df_mean.shape[0]):
  full_brightness_df_mean.iloc[j,:] = full_brightness_df_mean.iloc[j,:].where(full_brightness_df_mean.iloc[j,:] != 0, full_brightness_df_mean.loc[j,"brightness_mean"])
  
```

```python
any(full_brightness_df_mean["brightness_mean"] == 0)
```

```python
sum(full_brightness_df_mean["brightness_mean"] == 0)
```
